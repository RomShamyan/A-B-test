### Проект: вариант 1
### Задание 3
 В игре Plants & Gardens каждый месяц проводятся тематические события, ограниченные по времени. В них игроки могут получить уникальные предметы для сада и персонажей, дополнительные монеты или бонусы. Для получения награды требуется пройти ряд уровней за определенное время. С помощью каких метрик можно оценить результаты последнего прошедшего события?

 Предположим, в другом событии мы усложнили механику событий так, что при каждой неудачной попытке выполнения уровня игрок будет откатываться на несколько уровней назад. Изменится ли набор метрик оценки результата? Если да, то как?

#### Оценить результаты последнего прошедшего события можно с помощью следующих метрик:
- DAU/WAU/MAU — количество уникальных игоков, которые зашли в игру в течение суток/недели, месяца.
- Доля новых игоков — отношение числа новых игоков за какой-то промежуток времени ко всем игрокам за этот же промежуток времени
- stickness ratio = avg. DAU/MAU — коэффициент того, насколько часто игроки, в течение одного периода, "остаются" в продукте
- Retention rate (RR) — коэффициент удержания игроков. Метрика, которая показывает, как игроки возвращаются в продукт
- LTV (lifetime value) - пожизненная ценность клиента. Отражает, сколько в среднем приносит игрок денег в продукт до того, как покинет его
- ARPU (Average revenue per user) - средний доход с привлечённого игрока
- ARPPU (Average revenue per paying user) - cредний доход на платящего игрока
- Средняя продолжительность сессии (average session length) – среднее арифметическое длин всех сессий игроков в игре.
- Количество сеансов (сессий) на игрока — помогает понять как часто игроки возвращаются в игру, особенно если они не смоги пройти уровень
- Количестов игроков, которые прошли определенные уровни, а также длительность прохождения уровней.
- Количество игроков, получившие те или иные уникальные предметы, дополнительные монеты или бонусы.
    

В случае усложнения механики событий в игре помимо вышеперечисленных метрик нужно в первую очередь посмотреть на отток клиентов и метрику Churn rate (метрика, обратная retention).
Также дополнительно нужно проанализировать, какое количество игроков откатывается назад, как часто, на каких уровнях чаще всего застревают игроки и какое количество бросают в итоге из-за этого игру.
